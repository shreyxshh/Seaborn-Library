## Regression & Mixed Plots

Seaborn provides plots that combine **relationships** and **statistical models** or mix multiple plot types.

- `regplot()` / `lmplot()` → scatter + regression line  
- `jointplot()` → scatter with hist/KDE  
- `pairplot()` → multiple relationships in one go


```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

# regplot() – Scatter + Regression Line

```python
#used to plot linear regression plot 

sns.regplot(xlabel=tips['total_bill'], ylabel=tips['tip'])
```

# lmplot() – Regression with Facets

```python
#same plot as regplot

sns.lmplot(data=tips, x='total_bill', y='tip')
```

# jointplot() – Scatter + Histogram / KDE

```python
#combination of multiple plots 
#shows freqs and realtional plot
# freq = histogram relational/kind = multiple plot(shows what to be present on the centre of the plot)

sns.jointplot(x=tips['total_bill'], y=tips['tip'], kind='reg')
```

# pairplot() – Relationships Between Many Variables

```python
#used to show multiple relations at once for every numerical col

sns.pairplot(tips)
```

```python
#Heatmaps = used to show correlation btw numerical column only
 
sns.heatmap(tips.corr(numeric_only=True), annot=True)
#darker the  color = low corr
##lighter the color = high corr 
```

```python
iris = sns.load_dataset('iris')
sns.heatmap(iris.corr(numeric_only=True), annot=True)
```
